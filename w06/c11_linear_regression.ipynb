### Linear Regression
Given a dataset $X$ and $Y$, linear regression can be used to:

* Build a **predictive model** to predict future values of $X_i$ without a $y$ value.
* Model the strength of the relationship between each dependent variable $X_i$ and $y$
    - Sometimes not all $X_i$ will have a relationship with $y$
    - Need to figure out which $X_i$ contributes most information to determine $y$

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples=100, n_features=1, noise=0.1, random_state=1)
# print(X)
plt.scatter(X, y)
plt.show()
```

```python
# create and train the linear regression model:
model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)
plt.scatter(X, y)
plt.plot(X, y_pred, color='red') 
plt.show()
```

```python
# make regression with 2 features
X, y = make_regression(n_samples=100, n_features=2, noise=0.1, random_state=1)
#print(X)
X1 = X[:,1]
X1 = X1.reshape(100, 1)
#print(X1)
plt.scatter(X1, y)
plt.show()
```

```python
# create and train the linear regression model:
model = LinearRegression()
model.fit(X1, y)

y_pred = model.predict(X1)
plt.scatter(X1, y)
plt.plot(X1, y_pred, color='red') 
plt.show()
```

```python
# print the intercept and coefficient
print("intercept = ", model.intercept_)
print("coefficient = ", model.coef_)
```
