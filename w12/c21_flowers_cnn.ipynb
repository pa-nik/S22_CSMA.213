```python
from keras.models import load_model

model = load_model('flowers-cnn-model.h5')
model.summary()
```

```python
import matplotlib.pyplot as plt
from keras.preprocessing import image

img_path = 'sunflower.jpg'
#img_path = 'daisy.jpg'
img = image.load_img(img_path, target_size = (300, 300)) 
plt.imshow(img)
plt.show()
```

```python
img_array = image.img_to_array(img)
print(img_array.shape)
```

```python
img_array_4d = img_array.reshape(1,300,300,3)
print(img_array_4d.shape)
```

```python
prediction = model.predict(img_array_4d)
print(prediction)
```

```python
prediction_rounded = [ round(elem, 2) for elem in prediction[0] ]
print(prediction_rounded)
```

```python
import numpy as np

class_names = ['daisy', 'dandelion', 'roses', 'sunflowers', 'tulips']

index = np.argmax(prediction)
print('prediction:', class_names[index])
```
